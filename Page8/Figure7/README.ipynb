### Figure 7: Histogram of pTree's Size and Depth Distribution

The related experiment data is arranged as follow.

`pt_stat.csv` @ Each entry represents a JavaScript file and its pTree statistics
- `file_id`: The ID of the file
- `library`: The library name
- `file`: The file name
- `size`: The number of vertices in the pTree
- `depth`: The depth of the pTree

Following code paints the histogram of size distribution.

```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("./pt_stat.csv")
size_list = df['size'].tolist()
size_range_list = [0] * 5
for size in size_list:
    size_range_list[math.floor(math.log(size, 10))] += 1

plt.bar(x=range(5), height=size_range_list, width=0.5,
        color="#F5CCCC",
        edgecolor="#C66667")

label_list = ["[0,10)", "[10,100)", "[100,1000)", "[1000,10000)", "(10000,)"]
plt.xticks(range(5), label_list)
plt.xlabel("Size")
plt.ylabel("Number")

plt.rcParams['figure.figsize'] = [6.2, 3]
plt.show()
```

Following code paints the histogram of depth distribution.

```python
df= pd.read_csv("./pt_stat.csv")
depth_list = df['depth'].tolist()

plt.hist(x=depth_list, bins=5,
        color="#9FC5E9",
        edgecolor="#1c4586")

plt.xlabel("Depth")
plt.ylabel("Number")

plt.rcParams['figure.figsize'] = [3.5, 3]
plt.show()
```
